Notes and explainations for task 3 :

- for the missing basePay cells we filled it with the mean of the basePay because using the mean helps to preserve the central tendency of the data. Since the mean is the average value, it reflects the overall "center" of the distribution. and also When missing values are imputed with the mean, it minimally impacts the overall average of the dataset.

- for the missing OvertimePay cells I replaced it with 0


- for the missing OtherPay cells I replaced it with 0

- for the missing benefits cells we did the same as basePay



```python
#task 3 filling basePay cells

missing_values = df['BasePay'].isnull().sum()
print(f"Number of missing values in 'BasePay' column before cleaning: {missing_values}")


mean_BasePay = df['BasePay'].mean()
df['BasePay'].fillna(mean_BasePay, inplace=True)
print(f"Mean BasePay used for imputation: {mean_BasePay}")

missing_values_after_cleaning = df['BasePay'].isnull().sum()
print(f"Number of missing values in 'BasePay' column after cleaning: {missing_values_after_cleaning}")
```

```python
#task 3 filling OvertimePay cells

missing_values = df['OvertimePay'].isnull().sum()
print(f"Number of missing values in 'OvertimePay' column before cleaning: {missing_values}")


Fill_OvertimePay = '0'
df['OvertimePay'].fillna(Fill_OvertimePay, inplace=True)
print(f"Fill notes used for imputation: {Fill_OvertimePay}")

missing_values_after_cleaning = df['OvertimePay'].isnull().sum()
print(f"Number of missing values in 'OvertimePay' column after cleaning: {missing_values_after_cleaning}")
```

```python
#task 3 filling OtherPay cells

missing_values = df['OtherPay'].isnull().sum()
print(f"Number of missing values in 'OtherPay' column before cleaning: {missing_values}")


Fill_OtherPay = '0'
df['OtherPay'].fillna(Fill_OtherPay, inplace=True)
print(f"Fill OtherPay used for imputation: {Fill_OtherPay}")

missing_values_after_cleaning = df['OtherPay'].isnull().sum()
print(f"Number of missing values in 'OtherPay' column after cleaning: {missing_values_after_cleaning}")
```

```python
#task 3 filling Benefits cells

missing_values = df['Benefits'].isnull().sum()
print(f"Number of missing values in 'Benefits' column before cleaning: {missing_values}")


mean_Benefits = df['Benefits'].mean()
df['Benefits'].fillna(mean_Benefits, inplace=True)
print(f"Mean Benefits used for imputation: {mean_Benefits}")

missing_values_after_cleaning = df['Benefits'].isnull().sum()
print(f"Number of missing values in 'Benefits' column after cleaning: {missing_values_after_cleaning}")
```

```python
#task 3 filling Notes cells

missing_values = df['Notes'].isnull().sum()
print(f"Number of missing values in 'Notes' column before cleaning: {missing_values}")


Fill_Notes = 'no notes'
df['Notes'].fillna(Fill_Notes, inplace=True)
print(f"Fill notes used for imputation: {Fill_Notes}")

missing_values_after_cleaning = df['Notes'].isnull().sum()
print(f"Number of missing values in 'Notes' column after cleaning: {missing_values_after_cleaning}")
```

```python
#task 3 filling status cells

missing_values = df['Status'].isnull().sum()
print(f"Number of missing values in 'Status' column before cleaning: {missing_values}")


Fill_Status = 'no status'
df['Status'].fillna(Fill_Status, inplace=True)
print(f"Fill status used for imputation: {Fill_Status}")

missing_values_after_cleaning = df['Status'].isnull().sum()
print(f"Number of missing values in 'Status' column after cleaning: {missing_values_after_cleaning}")
```

- since there is no department column so am doing visualization 2 on the jobtitles so it will be a little messy

```python
#task 4


plt.figure(figsize=(10, 6))
plt.hist(df['TotalPay'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Salaries')
plt.xlabel('TotalPay')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# Visualization 2: Pie chart for the proportion of employees in different jobtitles
department_counts = df['JobTitle'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(department_counts, labels=department_counts.index, autopct='%1.1f%%', startangle=90, colors=['lightcoral', 'lightgreen', 'lightskyblue'])
plt.title('Proportion of Employees in Different Departments')
plt.show()
```

- since there is no department column so am doing the grouping by year and salary

```python

grouped_data = df.groupby('Year')['TotalPay'].mean().reset_index()

print("Grouped Data - Average Salary for Each Year:")
print(grouped_data)

# Visualization: Bar chart to compare average salaries across different years(since there is no department cell)
plt.figure(figsize=(10, 6))
plt.bar(grouped_data['Year'], grouped_data['TotalPay'], color='lightblue', edgecolor='black')
plt.title('Average Salary Across Different Years')
plt.xlabel('Year')
plt.ylabel('Average Salary')
plt.grid(axis='y')
plt.show()
```

- since there is no year of experience column so am doing the plot between salary and year columns

```python
correlation_coefficient = df['TotalPay'].corr(df['Year'])


print(f"Correlation Coefficient between Salary and Years of Experience: {correlation_coefficient}")

# Visualization: Scatter plot to visualize the relationship
plt.figure(figsize=(10, 6))
plt.scatter(df['Year'], df['TotalPay'], color='orange', edgecolor='black')
plt.title('Scatter Plot: Salary vs Years of Experience')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.grid(True)
plt.show()
```
